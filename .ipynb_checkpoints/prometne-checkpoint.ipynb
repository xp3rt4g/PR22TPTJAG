```python
import pandas as pd
import matplotlib.pyplot as plt

data2009 = pd.read_csv("promet-nesrece-podatki/pn2009.csv", sep=";", encoding="windows-1250")
data2010 = pd.read_csv("promet-nesrece-podatki/pn2010.csv", sep=";", encoding="windows-1250")
data2011 = pd.read_csv("promet-nesrece-podatki/pn2011.csv", sep=";", encoding="windows-1250")
data2012 = pd.read_csv("promet-nesrece-podatki/pn2012.csv", sep=";", encoding="windows-1250")
data2013 = pd.read_csv("promet-nesrece-podatki/pn2013.csv", sep=";", encoding="windows-1250")
data2014 = pd.read_csv("promet-nesrece-podatki/pn2014.csv", sep=";", encoding="windows-1250")
data2015 = pd.read_csv("promet-nesrece-podatki/pn2015.csv", sep=";", encoding="windows-1250")
data2016 = pd.read_csv("promet-nesrece-podatki/pn2016.csv", sep=";", encoding="windows-1250")
data2017 = pd.read_csv("promet-nesrece-podatki/pn2017.csv", sep=";", encoding="windows-1250")
data2018 = pd.read_csv("promet-nesrece-podatki/pn2018.csv", sep=";", encoding="windows-1250")
data2019 = pd.read_csv("promet-nesrece-podatki/pn2019.csv", sep=";", encoding="windows-1250")
data2020 = pd.read_csv("promet-nesrece-podatki/pn2020.csv", sep=";", encoding="windows-1250")
data2021 = pd.read_csv("promet-nesrece-podatki/pn2021.csv", sep=";", encoding="windows-1250")
data2022 = pd.read_csv("promet-nesrece-podatki/pn2022.csv", sep=";", encoding="windows-1250")

dataAll = pd.concat([data2009, data2010, data2011, data2012, data2013, data2014, data2015, data2016, data2017, data2018, data2019, data2020, data2021, data2022])
```

#### 2009

```python
dataAll.head()
```

```python
perClass = dataAll['KlasifikacijaNesrece'].value_counts()
print(perClass)

# Bar plot za št. nesreč na klasifikacijo
perClass.plot(kind='bar')
plt.title('Klasifikacija Nesreč (2009-2022)')
plt.xlabel('Klasifikacija nesreče')
plt.ylabel('Število nesreč')
plt.show()
```

```python
dataAll['DatumPN'] = pd.to_datetime(dataAll['DatumPN'], format='%d.%m.%Y')
perMonth = dataAll.groupby(dataAll['DatumPN'].dt.month)['DatumPN'].count()
print(perMonth)

# Bar plot za št. nesreč po mesecih
perMonth.plot(kind='bar')
plt.title('Število prometnih nesreč po mesecih (2009-2022)')
plt.xlabel('Mesec')
plt.ylabel('Število nesreč')
plt.show()
```

```python
dataAll['Starost'] = pd.to_numeric(dataAll['Starost'], errors='coerce').fillna(0).astype(int)
bins = [0, 21, 40, 60, float('inf')]
labels = ['0-21', '21-40', '40-60', '60+']
dataAll['age_group'] = pd.cut(dataAll['Starost'], bins=bins, labels=labels)
perAge = dataAll['age_group'].value_counts().sort_index()
print(perAge)

# Bar plot za št. nesreč glede na starostno skupino
perAge.plot(kind='bar')
plt.title('Število nesreč glede na starost (2009-2022)')
plt.xlabel('Starostna skupina')
plt.ylabel('Število nesreč')
plt.show()
```

```python
perRegion = data2019['TekstCesteNaselja'].value_counts()
print(perRegion[perRegion > 200])

#Bar plot za št. nesreč po regijah
perRegion[perRegion > 200].plot(kind='bar')
plt.title('Število nesreč glede na regijo (2009-2022)')
plt.xlabel('Regija')
plt.ylabel('Število nesreč')
plt.show()
```
